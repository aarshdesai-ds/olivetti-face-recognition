```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
import seaborn as sns
```

# üîç Loading Augmented Olivetti Dataset

In this section, we load the **Augmented Olivetti Faces Dataset**, which is stored as NumPy arrays:

- `augmented_faces.npy`: a 3D array of shape **(2000, 64, 64)** containing grayscale face images.
- `augmented_labels.npy`: a 1D array of shape **(2000,)** representing the subject label (person ID) corresponding to each image.

Each image is 64x64 pixels, and we'll need to reshape it for machine learning tasks.


```python
X = np.load("./data/augmented_faces.npy")
Y = np.load("./data/augmented_labels.npy")

print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

## üßπ Preprocessing the Data

Before we proceed with modeling and visualization, we need to prepare our dataset:

1. **Flatten the image data**:
   Each image of shape `(64, 64)` will be converted into a 1D vector of size `64*64 = 4096` so that it can be used in standard machine learning models.

2. **Create DataFrames**:
   We convert both the flattened image data and labels into pandas DataFrames for easier manipulation, analysis, and visualization.

3. **Merge features with labels**:
   Finally, we concatenate the image features and the labels into a single DataFrame, where each row corresponds to a single image sample.


```python
# Flatten the 64x64 images to 1D vectors
X_flat = X.reshape(X.shape[0], -1)

# Create DataFrames
features_df = pd.DataFrame(X_flat)
target_df = pd.DataFrame(Y, columns=["target"])

# Append target as a new column
features_df["target"] = target_df

```

### üìä Dataset Preview

Below is a preview of the merged dataset. It includes:

- **4096 columns** representing pixel intensity values of each image.
- **1 column** labeled `"label"` indicating the person ID for the corresponding face image.

This DataFrame will be used for visualization, dimensionality reduction (PCA), classification (KNN), and clustering (KMeans) in later steps.

```python
# Verify the number of rows and columns in Feature DataFrame
print('Number of rows and columns in the Features DataFrame:',features_df.shape)
features_df.head(10)
```

## üìä Target Label Distribution

Before training any machine learning models, it's essential to understand the distribution of our target labels.

In this dataset, the **`target` column** represents the person ID associated with each face image. Below, we:

- List all **unique labels** (i.e., unique individuals),
- Count how many images belong to each individual to verify **class balance**.

This step helps us ensure the dataset is suitable for classification and that each class is represented fairly.


```python
# Check the distribution of the labels in the target column.
print('Labels in the target column:',features_df['target'].unique())

print('\nDistribution of the labels in the target column :\n',features_df['target'].value_counts())
```

## üñºÔ∏è Visualize One Face Sample per Label

To better understand the data, it's helpful to visually inspect sample face images for each label (i.e., each person in the dataset).

### How it works:

- We **group the dataset by the `target` label**.
- A function `show_image(label)`:
  - Retrieves the **first face image** for the given label.
  - Reshapes the 4096-length pixel data back to a `64 x 64` image.
  - Displays it using `matplotlib`.

### Below:

We **loop through all 40 unique labels** and display one sample image for each person in the dataset.
This confirms:
- The data is correctly grouped by label.
- The faces are visually distinct.


```python
# Create the group object of the DataFrame
grouped_df = features_df.groupby(by='target')

# Define the function to visualise the images
def show_image(label_in_grouped_df):
  # Get the group for the input label
  group = grouped_df.get_group(label_in_grouped_df)
  # Get the row number of the first instance in the group
  row = group.index.values[0]
  # Get the data of the row number selected
  img = features_df.iloc[row,:-1].values.reshape(64,64)
  # Reshape the data into a 2D array of 64 x 64.
  plt.figure(figsize = (16, 5), dpi = 81)
  plt.title(f"Label {label_in_grouped_df}", fontsize = 15)
  plt.imshow(img,cmap='gray')
  plt.show()
  # Create the image
```

```python
# Call the function to create the images.
for i in range(40):
  show_image(i)
```

## üß™ Splitting the Dataset: Training and Testing Sets

To build and evaluate a predictive model, we need to divide the dataset into **training** and **testing** subsets.

### Here's what we do:

1. **Separate features and target**:
   - `X`: Contains the 4096 pixel intensity values (features)
   - `y`: Contains the `target` labels (person IDs)

2. **Split the dataset** using `train_test_split` from Scikit-Learn:
   - `test_size=0.3`: 30% of the data is reserved for testing
   - `random_state=42`: Ensures reproducibility of the split

3. **Print the shapes** of:
   - `X_train`, `X_test`: Training and testing feature matrices
   - `y_train`, `y_test`: Corresponding target label arrays

This step ensures that our model is trained on one portion of the data and evaluated on unseen samples to test gen


```python
features = list(features_df.columns[:-1])
X = features_df[features]
y = features_df['target']
# Split the feature and target arrays
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

# Print the shape of all the four variables i.e. 'X_train', 'X_test', 'y_train', and 'y_test'
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## ü§ñ Training a Linear SVM Model for Face Classification

To classify face images based on their visual features, we use a **Support Vector Classifier (SVC)** with a **linear kernel**, which is effective in high-dimensional spaces like our 4096-feature input.

### Steps:

1. **Model Initialization**  
   We create an instance of `SVC` from `sklearn.svm` and set `kernel='linear'` to use a linear decision boundary.

2. **Model Training**  
   Use the `.fit(X_train, y_train)` method to train the model on the training data.

3. **Training Accuracy Evaluation**  
   Call `.score(X_train, y_train)` to compute the accuracy of the model on the training set. This gives us a sense of how well the model is learning to separate the classes.

> Note: While this checks performance on training data, we will later evaluate it on the test set to check generalization.


```python
# Build a logistic regression model using the 'sklearn' module.
from sklearn.svm import SVC
# 1. Create the SVC model and pass 'kernel=linear' as input.
svc_model = SVC(kernel='linear')
# 2. Call the 'fit()' function with 'X_train' and 'y_train' as inputs.
svc_model.fit(X_train,y_train)
# 3. Call the 'score()' function with 'X_train' and 'y_train' as inputs to check the accuracy score of the model.
svc_model.score(X_train,y_train)
```

A score of **1.0** from svc_model.score(X_train, y_train) means that the Support Vector Machine classifier achieved 100% accuracy on the training set ‚Äî it predicted every training example's label correctly.

```python
# Check test accuracy
svc_model.score(X_test, y_test)
```

A test accuracy of **0.98** (or 98%) indicates that the model performs extremely well on unseen data.

## üß† Making Predictions and Analyzing Label Distribution

After training our SVM classifier, we now make predictions on both the **training** and **testing** datasets to inspect how the model is performing across all classes.

### üîç What We'll Do:

1. **Generate Predictions**  
   - Use the `.predict()` method on both `X_train` and `X_test`.

2. **Inspect Predicted Label Distribution**  
   - Count the occurrences of each predicted label using `.value_counts()`.  
   - This gives insight into how balanced the model's predictions are across the 40 unique face labels.

This analysis helps:
- Confirm that all labels are represented in predictions.
- Identify if the model is biased toward specific labels.
- Sanity check model output beyond just accuracy.
rtain labels.


```python
# Make predictions on the train dataset by using the 'predict()' function.
y_train_pred = pd.Series(svc_model.predict(X_train))
# Print the occurrence of each label computed in the predictions.
print('The occurrence of each label computed in the predictions : ',y_train_pred.value_counts(),sep='\n')
```

```python
# Make predictions on the test dataset.
y_test_pred = pd.Series(svc_model.predict(X_test))
# Print the occurrence of each label computed in the predictions.
print('The occurrence of each label computed in the predictions:',y_test_pred.value_counts(),sep='\n')
```

## üìâ Confusion Matrix & Classification Report (Training Set)

Now that we have the predictions for the training set, it's important to evaluate **how well the model performs per class** (i.e., per individual face label).

### What We Do:

1. **Confusion Matrix**  
   - A square matrix showing how often each label was predicted correctly (diagonal) vs. misclassified as another label (off-diagonal).
   - Visualized using a heatmap for better interpretability.

2. **Classification Report**  
   - Displays key metrics for each class:
     - **Precision**: How many selected items are relevant.
     - **Recall**: How many relevant items are selected.
     - **F1-score**: Harmonic mean of precision and recall.
     - **Support**: Number of actual occurrences of each class.

This deeper evaluation goes beyond overall accuracy and helps us understand the classifier's strengths and weaknesses per class.


```python
# Create the training confusion matrix DataFrame
conf_train_matrix = confusion_matrix(y_train,y_train_pred)
# Create the heatmap
plt.figure(figsize=(16,11),dpi=108)
sns.heatmap(conf_train_matrix,annot=True,fmt='g')
plt.show()
```

```python
# Print the classification report for the training predictions.
print(classification_report(y_train,y_train_pred))
```

## üß† Training Classification Report: Interpretation

The classification report for the training set shows **perfect performance** across all metrics:

### Key Highlights:

- **Precision, Recall, F1-score = 1.00** for all 40 classes (face labels)
- The model achieved **100% accuracy** on the training data.
- Support (number of samples per class) is relatively balanced, with values ranging between 29‚Äì40.

### üß† What This Means:

- ‚úÖ **Perfect Fit on Training Data**: The SVM model has successfully learned to classify every training image correctly.
- ‚ö†Ô∏è **Potential Overfitting**: While impressive, perfect training results can suggest the model may be overfitting‚Äîespecially if test performance is noticeably lower.
- üìå **Reassurance**: Since the test accuracy was also very high (98%), the model generalizes well and overfitting is likely not a concern in this case.

Next, we'll evaluate the same metrics on the **test set** to confirm generalization.


```python
# Create the testing confusion matrix DataFrame
conf_test_matrix = confusion_matrix(y_test,y_test_pred)
# Create the heatmap
plt.figure(figsize=(16,11),dpi=108)
sns.heatmap(conf_test_matrix,annot=True,fmt='g')
plt.show()
```

```python
# Print the classification report for the testing predictions.
print(classification_report(y_test,y_test_pred))
```

## üß™ Test Classification Report: Interpretation

After evaluating the model on the **test set**, we observe consistently high performance across all metrics.

### üîç Key Insights:

- **Overall Accuracy: 98%**
  - The model correctly predicted 98% of unseen face images.
  - This is only slightly lower than the training accuracy (100%), indicating **strong generalization**.

- **Precision, Recall, and F1-score**
  - Most classes (person IDs) have **perfect or near-perfect scores**.
  - A few labels (e.g., 2, 3, 7, 12, 20, 23, 25, 37, 39) show slight drops in either precision or recall.
  - These small dips are normal and expected in real-world testing and do not raise overfitting concerns.

- **Macro vs. Weighted Average**
  - **Macro Avg** treats all labels equally and scores 0.98.
  - **Weighted Avg** adjusts for support (sample count per class), and also scores 0.98 ‚Äî showing consistent performance even with slight class imbalance.

### ‚úÖ Summary:

- The model performs **exceptionally well** in predicting facial identities across a diverse and slightly imbalanced dataset.
- It strikes an excellent balance between **memorization** and **generalization**, confirming the effectiveness of a **linear SVM** on this high-dimensional facial data.

## üë• K-Nearest Neighbors (KNN) Classification

In this section, we build a **K-Nearest Neighbors (KNN)** classifier to predict face labels based on similarity to the training examples.

### Why KNN?

- Simple and intuitive classification algorithm
- No training phase‚Äîpredictions are made by comparing with neighbors
- Effective when similar samples lie close in feature space

We'll:
- Train a `KNeighborsClassifier`
- Evaluate accuracy on training and test sets
- Generate a classification report for detailed metrics


```python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

# Evaluate performance
print("KNN Training Accuracy:", knn_model.score(X_train, y_train))
print("KNN Testing Accuracy:", knn_model.score(X_test, y_test))

# Predictions
y_test_knn_pred = knn_model.predict(X_test)

# Confusion matrix + classification report
print("\nKNN Classification Report (Test Set):")
print(classification_report(y_test, y_test_knn_pred))
```

## üë• KNN Classification Report: Interpretation

After training and evaluating the **K-Nearest Neighbors (KNN)** classifier with `k=5`, we observe the following:

### üîπ Accuracy:
- **Training Accuracy**: 94.9%
- **Test Accuracy**: 86.2%
- Indicates that the model is learning well and **generalizes reasonably**, though it's not as powerful as the SVM.

### üîç Class-Level Insights:
- Several classes (e.g., 6, 13, 16, 26, 32, 33) achieved **perfect scores (precision, recall, F1 = 1.00)**.
- Some classes show **moderate to weak performance**:
  - Class `12`: Low recall (0.38) and F1-score (0.53)
  - Class `39`: F1-score = 0.60 due to lower recall
  - Indicates confusion with visually similar faces

### ‚öñÔ∏è Average Metrics:
- **Macro Average F1-score**: 0.86  
  (Averages scores for each class equally ‚Äî good for assessing per-class performance)
- **Weighted Average F1-score**: 0.86  
  (Takes class imbalance into account ‚Äî reflects good overall consistency)

### üß† Summary:
- KNN is **fast to implement** and performs **reasonably well**, but:
  - **Sensitive to local noise and class imbalance**
  - Not as robust or discriminative as SVM in high-dimensional facial data
- Still a solid baseline or secondary model

> ‚úÖ Next, you can explore dimensionality reduction and clustering (PCA + KMeans) to discover structure in the data without labels.


```python
# Reduce dimensionality for clustering
pca_kmeans = PCA(n_components=2)
X_pca_kmeans = pca_kmeans.fit_transform(X)

# Apply KMeans
kmeans = KMeans(n_clusters=40, random_state=42)
clusters = kmeans.fit_predict(X_pca_kmeans)

# Plot clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca_kmeans[:, 0], y=X_pca_kmeans[:, 1], hue=clusters, palette='tab20', legend=None)
plt.title("K-Means Clustering on PCA-Reduced Face Data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

## üß© Interpretation: K-Means Clustering on PCA-Reduced Face Data

The scatter plot visualizes K-Means clustering on the face dataset after reducing the original 4096-dimensional space to just 2 principal components using PCA. Each color represents a cluster of face images that K-Means identified as visually similar. Many clusters are clearly separated, indicating that the model has successfully grouped distinct identities. However, some overlaps exist, likely due to similar facial features or the limited representation capacity of 2D PCA. Overall, the clustering reveals meaningful structure in the data and confirms that unsupervised learning can uncover identity-like groupings even without labels.


## üìâ Principal Component Analysis (PCA)

PCA helps us visualize high-dimensional face image data by reducing it to fewer dimensions while preserving as much variance as possible.

### Goals:

- Understand the structure of face data
- Explore class separation in 2D
- Prepare data for unsupervised techniques like K-Means

We'll:
- Reduce data to 2D
- Visualize the true labels in 2D PCA space


```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot PCA with true labels
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=features_df['target'], palette='tab20', legend=None)
plt.title("PCA Projection of Augmented Face Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

## üìâ Interpretation: PCA Projection of Augmented Face Dataset

The PCA plot above visualizes the high-dimensional face images projected into 2 dimensions using Principal Component Analysis. Each dot represents a face image, and each color corresponds to a unique individual (true label). Despite reducing 4096 features down to just 2, PCA preserves a meaningful structure, showing distinguishable clusters for many individuals. Some classes overlap, which is expected given the compression, but overall the plot confirms that visual facial features carry strong identity signals. This representation helps us understand the global distribution of faces and supports downstream tasks like clustering and visualization.
